##### [1번 문제] 
파일 s.txt 내의 다음 내용과 함께 아래 문제의 코드를 제시하시오. 
>pig ham<br/>
>cat dog<br/>
>ham bird<br/>
>dog pig<br/>

```python
import os
```

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - 1) 파일 s.txt을 읽어서 각 라인에 있는 첫 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s1.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s1.txt에 있는 파일 내용은 다음과 같다.
>cat dog<br/>
>dog pig<br/>
>ham bird<br/>
>pig ham<br/>

```python
with open('s.txt', 'r') as f:
    mylist = []
    for line in f:
        lvalue = line.split()[0]
        rvalue = line.split()[1]
        mylist.append((lvalue, rvalue))
    mylist.sort()
    s = ""
    for i in mylist:
        s += "%s %s\n" % (i[0], i[1])
with open('s1.txt', 'w') as f: f.write(s)
with open('s1.txt', 'r') as f: print(f.read())
```

파일을 읽기 모드로 읽어서 단어들을 튜플 형태로 리스트에 저장한다. 그리고 첫 번째 단어를 기준으로 정렬한 후에 튜플을 다시 문자열로 언패킹한 후 s1.txt 파일에 저장한다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - 2) 파일 s.txt을 읽어서 각 라인에 있는 두 번째 단어(문자열) 자체들을 기준으로 라인별 정렬후 파일 s2.txt에 그 결과를 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s2.txt에 있는 파일 내용은 다음과 같다.
>ham bird<br/>
>cat dog<br/>
>pig ham<br/>
>dog pig<br/>

```python
with open('s.txt', 'r') as f:
    mylist = []
    for line in f:
        lvalue = line.split()[0]
        rvalue = line.split()[1]
        mylist.append((rvalue, lvalue))
    mylist.sort()
    s = ""
    for i in mylist:
        s += "%s %s\n" % (i[1], i[0])
with open('s2.txt', 'w') as f: f.write(s)
with open('s2.txt', 'r') as f: print(f.read())
```

파일을 읽기 모드로 읽어서 단어들을 튜플 형태로 리스트에 저장한다. 그리고 두 번째 단어를 기준으로 정렬한 후에 튜플을 다시 문자열로 언패킹한 후 s2.txt 파일에 저장한다.

&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  - 3) 파일 s.txt을 읽어서 각 라인들에 있는 모든 단어들을 순차적으로 다시 나열하되 각 라인에 세 개의 단어들이 오도록 하여 s3.txt에 기록하는 코드를 작성하시오. 즉, 프로그램 수행 후 s3.txt에 있는 파일 내용은 다음과 같다.
>pig ham cat<br/>
>dog ham bird<br/>
>dog pig<br/>

```python
with open('s.txt', 'r') as f:
    mylist = []
    for line in f: 
        for i in line.split(): mylist.append(i)
    s = ""
    for i in range(len(mylist)):
        if i % 3 == 0: s += '\n'
        s += "%s " % mylist[i]
with open('s3.txt', 'w') as f: f.write(s)
with open('s3.txt', 'r') as f: print(f.read())
```

파일을 읽기 모드로 읽어서 단어들을 모두 리스트에 저장한다. 그리고 문자열로 언패킹 하는데, 세 개씩 나누기 위해 현재 인덱스가 3으로 나누어 떨어질 경우 줄바꿈 문자를 첨가한다. 작업이 끝나면 해당 문자열을 s3.txt 파일에 저장한다.

  - 1) 프로그램 시작 후 다음과 같은 메시지가 출력되어 1, 2, 3 중 하나의 값을 입력 받도록 한다.

    - 2-1) 입력된 내용은 access.txt 라는 이름의 텍스트 파일 내에 저장된다. 
    - 2-2) access.txt 파일의 각 라인에는 가입된 회원 각각의 정보가 "[id]: [password], [name], [school]" 형태로 저장된다.
    - 2-3) 즉, 가입 회원이 10명이면 access.txt 파일 내에 라인 수도 정확히 10개이다.
    - 2-4) 암호화 방식은 sha 모듈을 활용한다. sha 모듈 활용 방법은 본 문제의 마지막에 제시된 sha 활용 예를 참고한다.
      - 즉, access.txt 파일 내에 password 정보는 암호화 되어 저장되어야 한다.
    - 2-5) 회원 정보를 입력 받을 때 id를 입력 받은 직후 access.txt를 확인하여 이미 존재하는 id가 입력되었다면 다음 메시지를 출력하고 id 정보를 다시 입력받는다.
      - Sorry, the entered ID is already used.
  - 3) 2를 선택하면 ID, Password의 2가지 정보를 입력받는 프로그램을 작성한다.
    - 3-1) 입력된 ID 정보가 access.txt에 존재하지 않으면 다음과 같은 메시지를 출력하고 다시 입력받는다.
      - Sorry, you are not a registered member.
    - 3-2) 입력된 ID가 올바르게 존재하지만 Password 정보가 access.txt 파일에 있는 정보와 불일치하면 다음과 같은 메시지를 출력하고 Password를 다시 입력받는다.
      - Sorry, the entered password is not correct.
      - 이 때에도 사용자가 입력한 Password 정보와 함께 sha 모듈이 활용되어야 한다.
    - 3-3) 입력된 ID와 Password가 모두 올바르면 다음과 같은 메시지를 출력한다.
      - Hello [name]!
      - 위 [name]에는 access.txt에 기록되어 있는 name 정보를 출력한다.
  - 4) 3을 선택하면 프로그램이 끝난다.
  - [참고] sha 모듈 활용 예

```python
import hashlib

def getMenu(): #선택한 값 받아오는 것을 함수화
    return int(input("""Welcome to Our Service

  1. Sign Up
  2. Sign In
  3. Quit
"""))

def getUID(s, idType): #0은 아이디, 1은 비밀번호, 2는 이름으로 설정하여 원하는 값을 불러오도록 만든 함수
    if idType == 0: return s.split(':')[0]
    else: return s.split(':')[1].split(',')[idType - 1].strip()

def encryptPW(s): #sha1 기법의 해싱
    return hashlib.sha1(s.encode()).hexdigest()

mMenu = getMenu() #메뉴 받아옴
while(1): #무한 반복
    if mMenu == 3: break #3번이면 프로그램을 종료함
    elif mMenu == 1:
        userID = input("아이디를 입력하세요 : ")
        userPW = input("비밀번호를 입력하세요 : ")
        userName = input("이름을 입력하세요 : ")
        userSchool = input("학교명을 입력하세요 : ")
        duplicated = False #아이디의 중복 여부 체크하는 변수
        if os.path.exists("access.txt"): #access.txt 파일이 존재하지 않으면 read 할 때 오류 발생하므로
            with open("access.txt", "r") as f: #파일을 read 모드로 오픈
                for line in f: #한 줄씩 읽으며
                    if userID == getUID(line, 0): #입력한 아이디가 이미 존재한다면
                        print("Sorry, the entered ID is already used.")
                        duplicated = True #중복 확인
                        break
            if duplicated: continue #중복이므로 입력을 다시 받기 위해 continue
        encrypted_userPW = encryptPW(userPW) #입력한 비밀번호 암호화
        with open("access.txt", "a") as f: #파일을 append 모드로 오픈
            f.write("%s: %s, %s, %s\n" % (userID, encrypted_userPW, userName, userSchool)) #파일에 작성
        print('\n' + '-' * 50)
    elif mMenu == 2:
        userID = input("아이디를 입력하세요 : ")
        userPW = input("비밀번호를 입력하세요 : ")
        if os.path.exists("access.txt"): #access.txt 파일이 존재하지 않으면 read 할 때 오류 발생하므로
            encrypted_userPW = encryptPW(userPW) #입력한 비밀번호 암호화
            isFound = False #아이디의 존재 여부 체크하는 변수
            with open("access.txt", "r") as f: #파일을 read 모드로 오픈
                for line in f: #한 줄씩 읽으며
                    if userID == getUID(line, 0) and encrypted_userPW == getUID(line, 1): #아이디와 비밀번호 모두 매칭되면
                        print("Hello %s!" % getUID(line, 2)) #이름 출력
                        isFound = True #존재 확인
                        break #반복문 탈출
                    elif userID == getUID(line, 0) and encrypted_userPW != getUID(line, 1): #아이디는 존재, 비밀번호는 틀렸을 시
                        while encrypted_userPW != getUID(line, 1): #비밀번호 맞게 입력할 때까지
                            print("Sorry, the entered password is not correct.")
                            encrypted_userPW = encryptPW(input("비밀번호를 입력하세요 : ")) #다시 비밀번호만 입력받음
                        print("Hello %s!" % getUID(line, 2)) #이름 출력
                        isFound = True #존재 확인
                        break #반복문 탈출
                if not isFound: print("Sorry, you are not a registered member.") #아이디를 발견하지 못할 시
        print('\n' + '-' * 50)
    mMenu = getMenu() #반복해서 메뉴 받아옴
```

##### [3번 문제]
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
  - 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
  > range(stop)<br/>
  > range(start, stop[, step])<br/>
  - 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
  - 3) frange 사용 예
    - 3-1) frange(0.5)
      - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
      - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
      - [2.2, 2.7, 3.2, 3.7]

```python
def frange(start, stop=None, step=0.1):
    if stop == None: 
        stop = start
        start = 0
    ListOfFrange = []
    while start < stop:
        ListOfFrange.append(start)
        start += step
    return ["%.1f" % i for i in ListOfFrange]
testcase = [frange(0.5), frange(1.0, 2.0), frange(2.2, 4.0, 0.5)]
for i in testcase: print(i)
```

처음 인자들에서 stop은 None 객체로, step은 0.1을 기본 인자로 주는 함수 frange를 선언한다. 그리고 만약 함수를 실행하였는데 stop 인자가 여전히 None 객체라면 인자를 하나만 받은 것이므로 start와 값을 교환한다. 이후 반복문을 통해 리스트에 값들을 순차적으로 추가하는 부분은 동일하다. 마지막에서 리스트 내부 값들의 정확성을 위해 리스트 내포 방식을 이용해 포매팅 시킨 후 반환한다.

##### [4번 문제] 
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오
- sum() 사용 예 
  - sum()
    - 0
  - sum(1, 2)
    - 3
  - sum(1, 2, 3, 4, 5)
    - 15
  - sum(1, 5, 7, 2, -10)
    - 5

```python
def sum(*args):
    SumOfArgs = 0
    for i in args: SumOfArgs += i
    return SumOfArgs
testcase = [sum(), sum(1,2), sum(1,2,3,4,5), sum(1,5,7,2,-10)]
for i in testcase: print(i)
```

가변인자를 받는 함수 sum을 선언해준다. 그리고 내부에서 언패킹하여 인자들의 총 합을 반복문으로 구한 후 반환한다.

##### [5번 문제] 
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
- 1) 다음에 제시되는 함수들을 모두 이용해야 한다. 
  - split
  - map
  - join
- 2) myinitial() 함수 사용 예
  - myinitial("as soon as possible")
    - ASAP

```python
def toupper(c):
    return c[0].upper()
def myinitial(s):
    return "".join(map(toupper, s.split()))
print(myinitial("as soon as possible"))
```

문자열을 인자로 받는 myinitial 함수를 하나 선언한다. 그리고 아래와 같은 순서로 작동한다.
1. 문자열을 공백 기준으로 분리한다.
2. map 함수를 통해 분리된 문자열이 toupper 함수에 의해 반환된다.
3. toupper 함수에서는 가장 처음 문자를 대문자로 변환시켜 반환한다.
4. 그렇게 나온 iterable한 map 객체를 join 함수를 통해 이어 붙인다.

##### [6번 문제]
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 

```python
def myfact(n):
    if n <= 1: return 1
    return n * myfact(n-1)
print(myfact(2900))
```

N의 팩토리얼은 1부터 N까지의 모든 정수의 곱이므로 1씩 감소해가며 곱하는 행위를 재귀적으로 구현하면 편리하다.

    - 반드시 pickle 모듈로 저장한 5개 이상의 pickle 파일들을 다시 5개 이상의 사전 객체로 로드 하는 코드가 추가되어야 함
  - 6) 간단한 검색엔진 코딩 (Like Google!!!)
    - 사용자에게 임의의 검색어 (하나 또는 여러 단어로 구성) 를 입력받으시오.
      - 예: 
        - 대한민국
        - 컴퓨터 공학
        - 맛있는 음식
        - 파이썬을 쉽게 공부하는 법
        - 한기대 장점을 알려주세요
    - 입력받은 검색어들에 대해서도 1)에서 제시하는 방법처럼 불용어 처리를 하여 정리
    - 입력 받은 검색어와 유사도(Similarity)가 높은 웹 사이트 기준으로 위 4)에서 미리 지정해 놓은 5개 이상의 웹 사이트 URL들을 일렬로 나열하여 출력하시오.
      - 검색어와 웹 사이트 간의 유사도는 본인이 스스로 정하시오.
      - 유사도를 정하는 최소한의 기준은 단어 출현 빈도를 기반으로 해야 하며, 이외의 본인이 생각하는 방안이 있으면 함께 사용해도 됨.
      - 유사도가 높은 웹 사이트가 상위에 출력되어야 함 (즉, 유사도 기준 내림 차순)
      - 유사도가 동일한 웹 사이트들에 대해서는 임의 배치함.
  - 7) [주의] 필수사항
    - 위에서 만든 검색엔진 코딩은 매우 간단한 것이라 부족한 점이 많이 존재한다.
    - 본인이 생각하기에 상업적인 완성도 높은 검색 로봇/엔진이 되려면 어떤 기능들이 추가적으로 구현되어야 할지 최소 1가지 이상 제시하시오.
