##### [2번 문제] 
다형성에 대해 설명하고 다형성을 보여주는 자신만의 파이썬 코드 예제를 제시하시오.

```python
class Duck():
    def quack(self):
        print("꽥꽥")
class WoodenDuck(Duck):
    def quack(self):
        print("나무오리는 울지 못해요.")
class Mallard(Duck):
    pass
duck = Duck()
duck.quack()
woodenduck = WoodenDuck();
woodenduck.quack()
mallard = Mallard();
mallard.quack()
```

__다형성(polymorphism)이라는__ 단어는 원래 '여러 개의 형태를 갖는다'라는 의미로 프로그래밍의 측면에서는 사용하는 객체에 따라 여러 가지로 동작하는 것을 가리킨다.<br>

##### [3번 문제] 

다음 각 요구사항 모두를 만족시키는 Counter 클래스를 코딩하시오 (정답을 각 요구사항별로 입력할 필요 없이 3번 문제에 대해 1개의 클래스 정의 코드를 제시하면 된다.)
  - 요구사항 1. 생성자에 count 값과 step 값을 인자로 받을 수 있다.
    - count: Counter 인스턴스가 지니는 초기 정수 값
    - step: Counter 인스턴스의 count가 증가되는 증분 (defalt 값: 1)

    > \>\>\> c = Counter(10) <br/>
    > \>\>\> d = Counter(10, 2) <br/>
    
  - 요구사항 2. 다음과 같이 Counter의 인스턴스를 출력을 해주는 \_\_str\_\_() 메소드를 Counter 클래스 내에 구현하시오.
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 10 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 10 <br/>
    
  - 요구사항 3. 다음과 같이 step에 주어진 증분만큼 count를 증가시키는 incr() 메소드를 Counter 클래스 내에 구현하시오.
    > \>\>\> c.incr()  <br/>
    > \>\>\> d.incr()  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 11 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 12 <br/>
    
  - 요구사항 4. Counter 클래스 내에 관련 메소드를 추가하여 인스턴스 객체를 직접 호출(call)할 수 있도록 하시오. 인스턴스 객체를 직접 호출했을 때에 내부적으로 incr() 메소드를 호출하는 방법으로 구현하시오. 
    > \>\>\> c()  <br/>
    > \>\>\> d()  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 12 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 14 <br/>
    
  - 요구사항 5. 다음과 같은 두 개의 산술 연산 (+, -)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> c = c + 5  <br/>
    > \>\>\> d = d - 5  <br/>
    > \>\>\> print(c)  <br/>
    > [Count (step: 1)] 17 <br/> 
    > \>\>\> print(d)  <br/>
    > [Count (step: 2)] 9 <br/>  
    
  - 요구사항 6. 다음과 같은 관계연산 (>, <, ==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.
    > \>\>\> print(c) > 10  <br/>
    > True <br/> 
    > \>\>\> print(d) > 10 <br/>
    > False <br/>
    > \>\>\> print(c) < 10  <br/>
    > False <br/> 
    > \>\>\> print(d) < 10 <br/>
    > True <br/>
    > \>\>\> print(c) == 17  <br/>
    > True <br/> 
    > \>\>\> print(d) != 9 <br/>
    > False <br/>

```python
class Counter():
    def __init__(self, count, step=1): #클래스의 생성자, step의 default는 1
        self.count = count #인스턴스 변수 count 설정
        self.step = step #인스턴스 변수 step 설정
    
    def __str__(self): #클래스의 출력자
        newstr = '[Count (step: %d)] %d' % (self.step, self.count) #요구 조건과 동일한 포맷으로 문자열 생성
        return newstr #생성한 문자열 반환
    
    def incr(self): #incr() 메소드
        self.count += self.step #step만큼 더함

    def __call__(self): #클래스 호출자
        self.count += self.step #step만큼 더함
        
    def __add__(self, other): #+오버라이드
        self.count += other #rvalue 인자만큼 더함
        return self #더한 결과값 반환
        
    def __sub__(self, other): #-오버라이드
        self.count -= other #rvalue 인자만큼 뺌
        return self #뺀 결과값 반환
    
    def __lt__(self, other): #<오버라이드
        return self.count < other #대소비교 결과값 반환
    
    def __le__(self, other): #<=오버라이드
        return self.count <= other #대소비교 결과값 반환
    
    def __gt__(self, other): #>오버라이드
        return self.count > other #대소비교 결과값 반환
    
    def __ge__(self, other): #>=오버라이드
        return self.count >= other #대소비교 결과값 반환
    
    def __eq__(self, other):#==오버라이드 
        return self.count == other #비교 결과값 반환
    
    def __ne__(self, other): #!=오버라이드
        return self.count != other #비교 결과값 반환
        
if __name__ == '__main__':
    c = Counter(10)
    d = Counter(10, 2)
    print(c)
    print(d)
    c.incr()
    d.incr()
    print(c)
    print(d)
    c()
    d()
    print(c)
    print(d)
    c = c + 5
    d = d - 5
    print(c)
    print(d)
    print(c > 10)
    print(d > 10)
    print(c < 10)
    print(d < 10)
    print(c == 17)
    print(d != 9)
```

##### [4번 문제] 

다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \_\_init\_\_(), \_\_str()\_\_(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.
  - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
#     s.eliminate_duplicate(s)
#     print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

\_\_init__과 \_\_str__은 특수 메소드들이다.<br>
__init은__ 생성자 메소드로 클래스명을 변수에 대입할 때 파라미터를 같이 넣어줄 경우 인스턴스 변수를 초기화할 수 있게 해준다. 아래 코드에서 s = MySet([1, 2, 2, 3])은 MySet 클래스를 생성할 때 init에서는 리스트를 받아 아이템을 추가하는 것을 볼 수 있다.<br>
__str은__ 출력할 때 포매팅을 만들어주는 메소드이다. 객체를 출력할 때 어떻게 출력할 것인지 임의로 조정할 수 있다. str 메소드를 보면 'MySet: {'을 먼저 출력하고 리스트의 아이템을 출력하도록 만들어져있다. 그에 따라 출력 결과를 보면 'MySet: {1 ,2 ,3}'로 설명한 내용과 동일함을 알 수 있다.<br>
__마지막으로__ @staticmethod는 말 그대로 정적 메소드이다. 클래스의 namespace에 묶여있지만 일반 메소드와 비슷하게 MySet.eliminate_duplicate()처럼 사용 가능하다. 하지만 인스턴스 메소드처럼 클래스 객체를 생성 후 사용할 수도 있다. 이를 테면 아래 코드 부분에서 주석 처리한 방법과 같다.

##### [5번 문제] 

4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오 
  - 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.
  - 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.
    > \>\>\> u = s | t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9} <br/> 
    
  - 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.
    > \>\>\> u = s & t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {2, 3} <br/>

  - 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다.
    > \>\>\> s = MySet([1, 2, 3])  <br/>
    > \>\>\> t = MySet([3, 4, 5])  <br/>
    > \>\>\> u = s - t  <br/>
    > \>\>\> print(u)  <br/>
    > MySet: {1, 2} <br/>

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    def __or__(self, other): #합집합
        newlist = self + other #두 리스트를 합친 후
        MySet.eliminate_duplicate(newlist) #중복된 값들을 제거한다.
        return newlist #완성된 리스트를 반환한다.
    
    def __and__(self, other): #교집합
        newlist = [] #새로운 집합 선언
        for i in self:
            if i in other: newlist.append(i) #겹치는 원소가 있다면 추가한다.
        MySet.eliminate_duplicate(newlist) #중복된 값들을 제거한다.
        return newlist #완성된 리스트를 반환한다.
    
    def __sub__(self, other): #차집합
        newlist = [] #새로운 집합 선언
        for i in self:
            if i not in other: newlist.append(i) #겹치지 않는 원소가 있다면 추가한다.
        MySet.eliminate_duplicate(newlist) #중복된 값들을 제거한다.
        return newlist #완성된 리스트를 반환한다.
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    #1
    u = s | t
    print(u)
    #2
    u = s & t
    print(u)
    #3
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u) 
```

    - 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.  
    

```python
#-*- coding:utf-8 -*-
import requests
import re
import string
import html
import math
```

```python
class SearchEngine():
    stopwords = []
    with open("stopwords.txt", "r", encoding='utf-8') as f:
        for line in f: stopwords.append(line.strip()) #불용어를 미리 리스트에 추가해둠
    
    def __init__(self, *args):
        self.urlList = [] #주소 리스트 초기화
        self.frequencyDictwithUrl = {} #주소 : (키워드 : 개수) 형태의 사전 초기화
        self.similarityDictwithUrl = {} #주소 : 유사도 형태의 사전 초기화
        for u in args: self.addUrl(u) #가변 인자로 받은 주소 추가
    
    def addUrl(self, url):
        self.urlList.append(url) #주소 리스트에 추가
        self.frequencyDictwithUrl[url] = self.countWords(url) #해당 주소의 단어 분석 결과를 추가
        
    def removeUrl(self, url):
        if url in self.urlList:
            self.urlList.remove(url) #주소 리스트에서도 지우고
            del self.frequencyDictwithUrl[url] #사전에서도 제거
        
    def listUrls(self):
        for u in self.urlList: print(u) #주소 출력
    
    def getWordsFrequency(self):
        if not self.urlList: return None #등록된 주소가 없으면 None 객체 반환
        frequencyDict = {} #단어 분석 결과들을 병합할 사전 생성
        for v in self.frequencyDictwithUrl.values():
            for key, value in v.items():
                if key in frequencyDict: frequencyDict[key] += value #단어가 이미 있다면 개수만큼 더함
                else: frequencyDict[key] = value #없다면 추가함
        return frequencyDict #병합된 빈도수 사전 반환
    
    def getMaxFreqencyWords(self):
        if not self.urlList: return None #등록된 주소가 없으면 None 객체 반환
        keywords = list(self.getWordsFrequency().items()) #병합된 빈도수 사전을 리스트로 변환 후
        keywords.sort(key=lambda k: k[1], reverse=True) #value를 기준으로 내림차순 정렬
        maxFrequency = keywords[0][1] #최대 빈도수 저장
        maxList = []
        for k in keywords:
            if k[1] == maxFrequency: maxList.append(k[0]) #최대 빈도수가 동일한게 있다면 추가
            else: break #아니면 탈출
        return maxList #최대 빈도 리스트 반환
    
    def searchUrlByWord(self, keyword):
        if not self.urlList: return None #등록된 주소가 없으면 None 객체 반환
        for u in self.urlList: self.similarityDictwithUrl[u] = self.similarityValidation(u, keyword) #유사도 분석
        sortedFrequency = list(self.similarityDictwithUrl.items()) #유사도 분석 결과를 리스트로 변환 후
        sortedFrequency.sort(key=lambda k: k[1], reverse=True) #value를 기준으로 내림차순 정렬
        maxFrequency = sortedFrequency[0][1] #최대 유사도 저장
        maxList = []
        for s in sortedFrequency:
            if s[1] == maxFrequency: maxList.append(s[0]) #최대 유사도가 동일한게 있다면 추가
            else: break #아니면 탈출
        return maxList #최대 유사도 리스트 반환
    
    def countWords(self, url): #단어 빈도 저장 메소드
        req = requests.get(url)
        source = re.sub(r'\n', ' ', html.unescape(req.text)) #모든 줄바꿈을 여백으로 대체
        source = re.sub(r'<(script|style).*?</\1>', '', source) #script와 style 제거
        source = re.sub(r'<!(--)?.*?-->', '', source) #html 주석 제거
        source = re.sub(r'</?.*?>', '<>', source) #HTML에서 모든 태그 제거
        line = source.split('<>') #임의로 정해둔 separator를 기준으로 파싱
        words = []
        resultDic = {}
        for i in line:
            if i.strip() == '': continue; #공백밖에 없으면 넘어감
            result = [r for r in i.strip().split()] #단어 사이의 whitespace를 기준으로 파싱
            for r in range(len(result)):
                for p in string.punctuation:
                     result[r] = result[r].replace(p, '') #punctuation 제거
                if result[r].strip() == '': continue #punctuation 제거했는데 공백이면 넘어감
                if result[r] in resultDic: resultDic[result[r]] += 1 #단어가 이미 저장되어 있으면 카운트 + 1
                else:
                    if result[r] not in self.stopwords: resultDic[result[r]] = 1 #불용어 제거하고 단어의 개수 저장
        return resultDic
                        
    def similarityValidation(self, url, key): #유사도 분석 메소드
        parsedkey = key.split()
        for i in range(len(parsedkey)): #불용어 제거 작업
            if parsedkey[i].lower() in self.stopwords: parsedkey[i] = '' #삭제하면 리스트의 크기가 변동되므로 공백으로 대체
        keywords = list(self.frequencyDictwithUrl[url].items()) #객체를 로드해서 리스트화 시킨 후
        maxword = 0 #총 단어의 개수를 저장할 변수
        tf = 0
        distance_cost = 0.0
        for k in keywords: maxword += k[1] #전체 단어의 개수 카운트
        for k in keywords: #비슷한 단어 세는 곳
            for pk in parsedkey: #입력받은 문자열에서
                if pk.strip() == '': continue #공백이면 넘어감
                pklower = pk.lower()
                klower = k[0].lower()
                if pklower in klower: tf += k[1] #검색어의 개수 카운트
                for i in range(len(pklower)):
                    distance = math.inf #거리 변수 초기화
                    for j in range(len(klower)):
                        if klower[j] == pklower[i]: distance = min(distance, abs(i - j)) #단어끼리 일치하는 가장 짧은 거리 찾아낸다
                    if distance != math.inf: distance_cost += pow(math.e, -distance) #함수로 매칭 값 반환
        if tf and distance_cost: similarity = math.log(tf * distance_cost)
            #검색어가 나온 횟수와 문자열 유사도를 곱한다. 너무 커질 수 있으니 로그 함수를 씌운다.
        else: similarity = 0 #검색어가 아예 존재하지 않으면 전체 가중치는 0
        return similarity
```

```python
w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com')
w3 = SearchEngine()
w1.addUrl('https://github.com')
w3.addUrl('http://stackoverflow.com')
w1.removeUrl('https://www.cnn.com')
w2.removeUrl('http://stackoverflow.com')
w1.listUrls()
print(w1.getWordsFrequency())
print(w1.getMaxFreqencyWords())
print(w1.searchUrlByWord("news"))
```

##### [8번 문제] 

위 7번 문제에서 정의한 SearchEngine 클래스를 상속하여 SearchEngineWithOrderedWebWords 클래스를 정의하고 슈퍼클래스에 정의된 getWordsFrequency() 메소드를 ***오버라이드***하여 단어 출현 빈도를 내림 차순으로 정렬하여 리스트로 출력하시오.
  - 리스트 내의 각 원소는 단어와 빈도를 쌍으로 지니는 튜플이다.
  - getWordFrequence() 메소드에 reverse 라는 인자를 만들고 true 또는 false를 인자로 받을 수 있도록 한다.
    - reverse 인자의 디폴트 값은 false 이며, 기본적으로 내림차순으로 정렬한다.
    - reverse 인자에 true를 넣으면 오름차순으로 정렬한다.
    
  > \>\>\> w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
    
  > \>\>\> w4.getWordsFrequency()<br/>
  > [('site', 12), ('science', 11), ('hello', 8), ('world', 2), ('program', 1), ('python', 1)]
    
  > \>\>\> w4.getWordsFrequency(reverse=true)<br/>
  > [('program', 1), ('python', 1), ('world', 2), ('hello', 8), ('science', 11), ('site', 12)]

```python
class SearchEngineWithOrderedWebWords(SearchEngine): #SearchEngine을 상속으로 하는 클래스 선언
    def __iter__(self): #반복자
        return iter(self.global_keywords) #리스트를 반복자로 반환
    
    def getWordsFrequency(self, reverse=False):
        if not self.urlList: return None #등록된 주소가 없으면 None 객체 반환
        frequencyDict = {} #단어 분석 결과들을 병합할 사전 생성
        for v in self.frequencyDictwithUrl.values():
            for key, value in v.items():
                if key in frequencyDict: frequencyDict[key] += value #단어가 이미 있다면 개수만큼 더함
                else: frequencyDict[key] = value #없다면 추가함
        self.global_keywords = list(frequencyDict.items()) #병합된 빈도수 사전을 리스트로 변환 후
        self.global_keywords.sort(key=lambda k: k[1], reverse=not reverse) #정렬
        self.global_keywords = list(self.global_keywords) #리스트로 변환
        return self.global_keywords #정렬된 리스트 반환
```

```python
w4 = SearchEngineWithOrderedWebWords('http://www.times.com', 'https://www.amazon.com', 'https://github.com')
print(w4.getWordsFrequency())
# print()
# print(w4.getWordsFrequency(reverse=True))
```
