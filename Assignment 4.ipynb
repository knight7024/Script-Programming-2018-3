>cat dog<br/>
>dog pig<br/>
>ham bird<br/>
>pig ham<br/>

```python
with open('s.txt', 'r') as f:
    mylist = []
    for line in f:
        lvalue = line.split()[0]
        rvalue = line.split()[1]
        mylist.append((lvalue, rvalue))
    mylist.sort()
    s = ""
    for i in mylist:
        s += "%s %s\n" % (i[0], i[1])
with open('s1.txt', 'w') as f: f.write(s)
with open('s1.txt', 'r') as f: print(f.read())
```

```python
import hashlib

def getMenu(): #선택한 값 받아오는 것을 함수화
    return int(input("""Welcome to Our Service

  1. Sign Up
  2. Sign In
  3. Quit
"""))

def getUID(s, idType): #0은 아이디, 1은 비밀번호, 2는 이름으로 설정하여 원하는 값을 불러오도록 만든 함수
    if idType == 0: return s.split(':')[0]
    else: return s.split(':')[1].split(',')[idType - 1].strip()

def encryptPW(s): #sha1 기법의 해싱
    return hashlib.sha1(s.encode()).hexdigest()

mMenu = getMenu() #메뉴 받아옴
while(1): #무한 반복
    if mMenu == 3: break #3번이면 프로그램을 종료함
    elif mMenu == 1:
        userID = input("아이디를 입력하세요 : ")
        userPW = input("비밀번호를 입력하세요 : ")
        userName = input("이름을 입력하세요 : ")
        userSchool = input("학교명을 입력하세요 : ")
        duplicated = False #아이디의 중복 여부 체크하는 변수
        if os.path.exists("access.txt"): #access.txt 파일이 존재하지 않으면 read 할 때 오류 발생하므로
            with open("access.txt", "r") as f: #파일을 read 모드로 오픈
                for line in f: #한 줄씩 읽으며
                    if userID == getUID(line, 0): #입력한 아이디가 이미 존재한다면
                        print("Sorry, the entered ID is already used.")
                        duplicated = True #중복 확인
                        break
            if duplicated: continue #중복이므로 입력을 다시 받기 위해 continue
        encrypted_userPW = encryptPW(userPW) #입력한 비밀번호 암호화
        with open("access.txt", "a") as f: #파일을 append 모드로 오픈
            f.write("%s: %s, %s, %s\n" % (userID, encrypted_userPW, userName, userSchool)) #파일에 작성
        print('\n' + '-' * 50)
    elif mMenu == 2:
        userID = input("아이디를 입력하세요 : ")
        userPW = input("비밀번호를 입력하세요 : ")
        if os.path.exists("access.txt"): #access.txt 파일이 존재하지 않으면 read 할 때 오류 발생하므로
            encrypted_userPW = encryptPW(userPW) #입력한 비밀번호 암호화
            isFound = False #아이디의 존재 여부 체크하는 변수
            with open("access.txt", "r") as f: #파일을 read 모드로 오픈
                for line in f: #한 줄씩 읽으며
                    if userID == getUID(line, 0) and encrypted_userPW == getUID(line, 1): #아이디와 비밀번호 모두 매칭되면
                        print("Hello %s!" % getUID(line, 2)) #이름 출력
                        isFound = True #존재 확인
                        break #반복문 탈출
                    elif userID == getUID(line, 0) and encrypted_userPW != getUID(line, 1): #아이디는 존재, 비밀번호는 틀렸을 시
                        while encrypted_userPW != getUID(line, 1): #비밀번호 맞게 입력할 때까지
                            print("Sorry, the entered password is not correct.")
                            encrypted_userPW = encryptPW(input("비밀번호를 입력하세요 : ")) #다시 비밀번호만 입력받음
                        print("Hello %s!" % getUID(line, 2)) #이름 출력
                        isFound = True #존재 확인
                        break #반복문 탈출
                if not isFound: print("Sorry, you are not a registered member.") #아이디를 발견하지 못할 시
        print('\n' + '-' * 50)
    mMenu = getMenu() #반복해서 메뉴 받아옴
```

##### [3번 문제]
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.
  - 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.
  > range(stop)<br/>
  > range(start, stop[, step])<br/>
  - 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다. 
  - 3) frange 사용 예
    - 3-1) frange(0.5)
      - [0.0, 0.1, 0.2, 0.3, 0.4]
    - 3-2) frange(1.0, 2.0)
      - [1.0, 1.1, 1.2, 1.3, 1.4, 1.5, 1.6, 1.7, 1.8, 1.9]
    - 3-3) frange(2.2, 4.0, 0.5)
      - [2.2, 2.7, 3.2, 3.7]

```python
def frange(start, stop=None, step=0.1):
    if stop == None: 
        stop = start
        start = 0
    ListOfFrange = []
    while start < stop:
        ListOfFrange.append(start)
        start += step
    return ["%.1f" % i for i in ListOfFrange]
testcase = [frange(0.5), frange(1.0, 2.0), frange(2.2, 4.0, 0.5)]
for i in testcase: print(i)
```

처음 인자들에서 stop은 None 객체로, step은 0.1을 기본 인자로 주는 함수 frange를 선언한다. 그리고 만약 함수를 실행하였는데 stop 인자가 여전히 None 객체라면 인자를 하나만 받은 것이므로 start와 값을 교환한다. 이후 반복문을 통해 리스트에 값들을 순차적으로 추가하는 부분은 동일하다. 마지막에서 리스트 내부 값들의 정확성을 위해 리스트 내포 방식을 이용해 포매팅 시킨 후 반환한다.

##### [4번 문제] 
가변인수를 받는 함수 sum()을 다음과 같은 조건을 참고하여 구현하시오
- sum() 사용 예 
  - sum()
    - 0
  - sum(1, 2)
    - 3
  - sum(1, 2, 3, 4, 5)
    - 15
  - sum(1, 5, 7, 2, -10)
    - 5

```python
def sum(*args):
    SumOfArgs = 0
    for i in args: SumOfArgs += i
    return SumOfArgs
testcase = [sum(), sum(1,2), sum(1,2,3,4,5), sum(1,5,7,2,-10)]
for i in testcase: print(i)
```

가변인자를 받는 함수 sum을 선언해준다. 그리고 내부에서 언패킹하여 인자들의 총 합을 반복문으로 구한 후 반환한다.

##### [5번 문제] 
여러 단어로 이루어진 문자열을 입력받아 각 단어의 첫글자로 이루어진 단어를 대문자로 출력하는 myinitial() 함수를 다음 조건을 참고하여 작성하시오.
- 1) 다음에 제시되는 함수들을 모두 이용해야 한다. 
  - split
  - map
  - join
- 2) myinitial() 함수 사용 예
  - myinitial("as soon as possible")
    - ASAP

```python
def toupper(c):
    return c[0].upper()
def myinitial(s):
    return "".join(map(toupper, s.split()))
print(myinitial("as soon as possible"))
```

문자열을 인자로 받는 myinitial 함수를 하나 선언한다. 그리고 아래와 같은 순서로 작동한다.
1. 문자열을 공백 기준으로 분리한다.
2. map 함수를 통해 분리된 문자열이 toupper 함수에 의해 반환된다.
3. toupper 함수에서는 가장 처음 문자를 대문자로 변환시켜 반환한다.
4. 그렇게 나온 iterable한 map 객체를 join 함수를 통해 이어 붙인다.

##### [6번 문제]
음이 아닌 정수 n를 입력받으면 n! (factorial)을 계산하는 myfact() 함수를 재귀적 함수로 구현하시오.
- [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다. 

```python
def myfact(n):
    if n <= 1: return 1
    return n * myfact(n-1)
print(myfact(2900))
```

N의 팩토리얼은 1부터 N까지의 모든 정수의 곱이므로 1씩 감소해가며 곱하는 행위를 재귀적으로 구현하면 편리하다.

```python
#-*- coding:utf-8 -*-
import requests
import re
import string
import html
import pickle
# import math
stopwords = []
with open("stopwords.txt", "r", encoding='utf-8') as f:
    for line in f: stopwords.append(line.strip()) #불용어를 미리 리스트에 추가해둠
```

```python
def saveHTML(url):
    with open("%s.html" % url.split('//')[1], "w", encoding='utf-8') as f:
        f.write(requests.get(url).text) #HTML 저장
```

```python
def pickKeyword(urls):
    for u in urls:
        with open('%s.words.frequency' % u.split('//')[1], "rb") as f:
            keywords = list(pickle.load(f).items()) #객체를 로드해서 리스트화 시킨 후
            keywords.sort(key=lambda k: k[1], reverse=True) #value를 기준으로 내림차순 정렬
            print("%s 에서 가장 많이 출현한 단어 3개 :" % u)
            for i in range(3):
                print("(%s, %d번)" % (keywords[i][0], keywords[i][1]), end=', ') #단어 3개 출력
            print()
```

```python
# def distanceMatching(d):
#     return pow(math.e, -1 * abs(d)) # y=e^(-|d|) (단, d는 문자와 문자 사이의 거리)
```

```python
def similarityValidation(urls, key):
    parsedkey = key.split()
    for i in range(len(parsedkey)): #불용어 제거 작업
        if parsedkey[i].lower() in stopwords: del parsedkey[i]
    print("웹사이트에서의 유사도 판별 :")
    similarity = {}
    for u in urls:
        with open('%s.words.frequency' % u.split('//')[1], "rb") as f:
            datalist = pickle.load(f)
            keywords = list(datalist.items()) #객체를 로드해서 리스트화 시킨 후
            maxword = 0 #총 단어의 개수를 저장할 변수
            tf = 0
            cost = 0.0
            for k in keywords:
                if key in k[0].lower(): tf += k[1] #검색어의 개수 카운트
                maxword += k[1] #단어의 개수 카운트
            cost_per_keyword = 100 / maxword #글자당 가중치, 이하 유사 가중치
            for k in keywords: #비슷한 단어 세는 곳
                sMatching = set() #몇 개의 글자가 포함되어 있는지 매칭시킬 집합 선언
                for one in parsedkey: #입력받은 문자열에서
                    for c in one: #한 단어에서도 한 글자씩
                        if c.lower() in k[0]: sMatching.add(c.lower()) #중복되지 않는 글자의 개수 세기 위해
                cost += cost_per_keyword * len(sMatching) #글자당 가중치 * 포함된 글자
            if cost != 0: similarity[u] = tf / cost #검색어가 나온 횟수를 유사 가중치로 나눔
            else: similarity[u] = 0 #검색어가 아예 존재하지 않으면 전체 가중치는 0 
    sortedSimilarity = list(similarity.items())
    sortedSimilarity.sort(key=lambda k: k[1], reverse=True) #가중치 별로 정렬 후 출력
    for i in sortedSimilarity:
        print("%s : %f" % (i[0], i[1]))
```

```python
def countWords(url):
    req = requests.get(url)
    source = re.sub(r'\n', ' ', html.unescape(req.text)) #모든 줄바꿈을 여백으로 대체
#     source = re.sub(r'<(script|style).*?</\1>', '', source) #script와 style 제거
    source = re.sub(r'<!(--)?.*?-->', '', source) #html 주석 제거
    source = re.sub(r'</?.*?>', '<>', source) #HTML에서 모든 태그 제거
    line = source.split('<>') #임의로 정해둔 separator를 기준으로 파싱
    words = []
    resultDic = {}
    for i in line:
        if i.strip() == '': continue; #공백밖에 없으면 넘어감
        result = [r for r in i.strip().split()] #단어 사이의 whitespace를 기준으로 파싱
        for r in range(len(result)):
            for p in string.punctuation:
                 result[r] = result[r].replace(p, '') #punctuation 제거
            if result[r].strip() == '': continue #punctuation 제거했는데 공백이면 넘어감
            if result[r] in resultDic: resultDic[result[r]] += 1 #단어가 이미 저장되어 있으면 카운트 + 1
            else: resultDic[result[r]] = 1 #단어의 개수 저장
    for k in resultDic.keys(): #불용어 제거 작업
        if k.lower() in stopwords: resultDic[k] = -1 
            #사전에 불용어가 발견되면 개수를 음수로 바꿔 무의미한 값으로 만든다.
            #Why? 사전은 반복문 도중 크기가 바뀌어서는 안되기 때문
    with open("%s.words.frequency" % url.split('//')[1], "wb") as f: pickle.dump(resultDic, f) #피클로 객체 저장
    #return resultDic
urls = ['http://cse.koreatech.ac.kr', 'https://www.koreatech.ac.kr', 'https://news.naver.com', 'https://www.billboard.com', 'https://bbc.com']
for u in urls:
    saveHTML(u)
    countWords(u)
    #     print(countWords(u), end='\n\n')
```

#### 유사도 판별 알고리즘 상세 설명
유사도를 판별하기 위해 중요한 것은 단어 두 개가 얼마나 비슷한가와 단어가 완전히 일치하냐이다. 얼마나 비슷한지를 판별하는 방법으로 두 가지를 생각해보았다. <br>
* 가나다 5개, 나다라 3개, 다라마 9개, 라마바 8개로 총 25개의 단어가 있다고 가정하면 글자 하나당 100 / 25로 4%를 부여한다고 생각한다. 검색어로 '가라'가 들어오면 가나다에서 가라와 일치하는 단어는 1개로 4 / 3(글자의 길이)%이고 나머지도 모두 같으므로 4%라는 결과가 나온다.
* museum이라는 단어가 있는데 검색어로 mum이 들어왔다. 앞에 mu까지는 둘이 일치하고 mum에서 마지막 m과 museum에서 마지막 m은 거리가 3이다. 거리가 작을 수록 높은 유사도가 있어야하고 멀어질 수록 서서히 작아져야 하므로 y=e^(-|d|)라는 식을 떠올렸다. 이 식을 응용하여 유사도 가중치를 부여해볼까 했지만 구체화에 실패하여 미숙하지만 1번 방안을 채택하였다.


마지막으로 완전히 일치하는 횟수를로 카운트한 다음, (일치하는 횟수 / 유사 개수) 를 식으로 해서 유사도를 계산한다. 일치 횟수는 가장 중요하므로 분자에 배치하고 유사 개수는 페이지 내에 단순히 많은 단어가 있으면 계속해서 커질 우려가 있어 분모에 배치하였다.
